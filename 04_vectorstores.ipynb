# 1. FAISS
 - Facebook AI Similarity Search

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader = TextLoader('Data/speech.txt')
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=30)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embedding = OllamaEmbeddings(model='gemma:2b')
db = FAISS.from_documents(docs,embedding)
db
```

```python
## quering
query = "What does the speaker believe is the main reason the United States should enter the war?"
docs = db.similarity_search(query)
docs
```

```python
print(docs[0].page_content)
```

### Retriever:
We can also convert the vectorestore into a Retriever Class. This allows us to easily use it in other LangChain methods,which largely work with retrievers

```python
retriever = db.as_retriever()
docs = retriever.invoke(query)
```

```python
docs[0].page_content
```

#### Similarity Search with score
There are some FAISS specific methods.One of them is similarity_search_with_score,which allows you to return not only the documents but also the distance score of the query to them.The returned distance score is L2 distance.Therefore,a lower score is better.

```python
docs_ans_sscore = db.similarity_search_with_score(query)
docs_ans_sscore
```

```python
embedding_vector = embedding.embed_query(query)
embedding_vector
```

```python
docs_vscore = db.similarity_search_by_vector(embedding_vector)
docs_vscore
```

#### Saving and Loading

```python
db.save_local('faiss_index')
```

```python
new_db = FAISS.load_local('faiss_index',embedding,allow_dangerous_deserialization=True)
```

```python
docs = new_db.similarity_search(query)
```

```python
docs
```

# 2. CHROMADB

Chroma is a AI-native open source vector database focused on developer productivity and happiness.Chroma is licensed under Apache 2.0

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings
```

```python
loader = TextLoader('Data/speech.txt')
data = loader.load()
data
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=0)
splits = text_splitter.split_documents(data)
splits
```

```python
embedding = OllamaEmbeddings(model='gemma:2b')
vectordb = Chroma.from_documents(splits,embedding)
```

```python
vectordb
```

```python
## quering
query = "What does the speaker believe is the main reason the United States should enter the war?"
docs = vectordb.similarity_search(query)
docs[0].page_content
```

```python
# save to the disk
vectordb = Chroma.from_documents(documents=splits,embedding=embedding,persist_directory='./chroma_db')
```

```python
# Load from disk
db2 = Chroma(persist_directory="./chroma_db",embedding_function=embedding)
docs = db2.similarity_search(query)
docs[0].page_content
```

```python
docs = db2.similarity_search_with_score(query)
docs
```

```python
retriever = vectordb.as_retriever()
docs = retriever.invoke(query)
docs[0].page_content
```
