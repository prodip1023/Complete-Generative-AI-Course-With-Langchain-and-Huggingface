# 1.  WebBaseLoader
This covers how to use WebBaseLoader to load all text from HTML webpages into a document format that we can use downstream.

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4
```

```python
loader = WebBaseLoader(web_paths=("https://en.wikipedia.org/wiki/Transformer",),
                      bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                          class_ = ('mw-page-title-main','infobox')
                      )) )
```

```python
web_data = loader.load()
```

```python
web_data[0].metadata
```

# 10. DuckDB
DuckDB is an in-process SQL OLAP database management system.

```python
from langchain_community.document_loaders import DuckDBLoader
```

```python
loader = DuckDBLoader("SELECT * FROM read_csv_auto('Data/listings.csv')")
data = loader.load()
print(data[0].page_content)
```

```python
loader = DuckDBLoader("SELECT * FROM read_csv_auto('Data/listings.csv')",
                     page_content_columns = ['room_type'],
                      metadata_columns = ['price'],
                     )
data = loader.load()
```

```python
for i in range(len(data)):
    print(data[i].page_content)
    print(data[i].metadata)
```

# 17. PySpark
PySpark is a Python API for Apache Spark. It allows you to write Spark applications in Python.
https://python.langchain.com/docs/integrations/document_loaders/pyspark_dataframe/
